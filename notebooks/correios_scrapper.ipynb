# Street Scrapper

Collect Brazilian street names from Correios using Selenium.

The goal is to create a NLP spellcheck model trained with a corpus specific to street and neighborhood names. Correios has the best updated repository of street and neighborhood names, including all the Brazilian territory. By training a model with this specific text corpus, we can enhance the quality of the spatial information provided in the DATASUS databases and improve geocoding.

## Scrape data with Selenium

```python
url = "https://buscacepinter.correios.com.br/app/endereco/index.php?t"
```

```python
# --> Open driver
driver = webdriver.Firefox()
driver.get(url)

# --> Search adress based on the string provided and press ENTER
endereco_form = driver.find_element(By.ID, 'endereco')
endereco_form.send_keys("ALDEOTA, FORTALEZA", Keys.ENTER)
```

```python
npage, max_page = 1, 20
rows_total = []
while npage<=max_page:
    # --> Extract table of addresses
    try:
        tbody_block = WebDriverWait(driver, 20.0).until(EC.presence_of_element_located((By.TAG_NAME, 'tbody')))
    except:
        print("<tbody> not located")
        driver.close()
        
    sleep_time = np.random.uniform()*2
    time.sleep(sleep_time)
    
    # --> Extract HTML table and read structured data into list
    html_table = tbody_block.get_attribute("outerHTML")
    bs4_parser = BeautifulSoup(html_table, "html.parser")
    tb_rows = bs4_parser.find_all("tr")
    
    rows = []
    for current_row in tb_rows:
        cell_blocks = current_row.find_all("td")
        rows.append( [ cell.get_text() for cell in cell_blocks ] )
    rows_total += rows    
    # --> Structure the data
    #fdata = [
    #    {"QUERY": "Ceará", "NUMERO_PAGINA": npage, "LOGRADOURO/NOME": n[0],
    #    "BAIRRO/DISTRITO": n[1], "LOCALIDADE/UF": n[2],
    #    "CEP": n[3]} for n in rows
    #]
    
    # --> Insert data in DUCKDB database
    #ins = logr_correios_ceara.insert()
    #with engine.connect() as conn:
    #    rp = conn.execute(ins, fdata)
    #    conn.commit()
        
    print(f"Page {npage}: {len(rows)} extracted")
    print(f"Last row extracted: {fdata[-1]}")
        
    # --> Find next button
    botao_proximo = WebDriverWait(driver, 20.0).until(EC.presence_of_element_located((By.XPATH, '//a[@href="javascript:pesquisarProximo()"]')))
    try:
        botao_proximo.click()
        npage+=1
    except:
        print("Next buttom not iteractable - End of search")
        driver.close()
        break
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(os.path.join(os.environ["HOMEPATH"], "Downloads", "TMI TOTAL.xlsx"))
xdata = df['Ano']

df1 = df.iloc[:, 1:]
df1.columns = [ f"{x}".replace("TMI ", "") for x in df1.columns ]
df1 = df1.rename({"Ceara": "Ceará"}, axis=1)
```

```python
fig, ax = plt.subplots(1, figsize=(8,5))

csfont = {'fontname':'Comic Sans MS'}
hfont = {'fontname':'Helvetica'}

ax.spines["top"].set_linewidth(0)
ax.spines["right"].set_linewidth(0)
ax.spines["left"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
ax.spines['left'].set_position(('outward', 12))
ax.spines['bottom'].set_position(('outward', 10))

ax.tick_params(width=1.5, labelsize=11)

ax.set_xticks(xdata.values)
ax.set_xticklabels([ f'{x}' for x in xdata ], rotation=0)
ax.set_xlim([2010, 2021])
ax.set_ylim([8, 20])
ax.grid(alpha=0.1)

ax.set_xlabel("Ano", fontsize=14, weight='bold', labelpad=7)
ax.set_ylabel("Taxa de Mortalidade Infantil", fontsize=12, weight='bold', labelpad=9)

# plot

c = ['black', 'tab:orange', 'tab:green', 'tab:blue', 'tab:purple', 'tab:red']
th = [2., 2., 2., 2.25, 2., 2.]
for index, name in enumerate(df1.columns):
    ax.plot(xdata.values, df1[f'{name}'].values, color=c[index], label=name, lw=th[index], ls='-')

#ax.plot(xdata.values, df1['Ceara'].values)
#ax.plot(xdata.values, df1['Litoral Leste'].values)
#ax.plot(xdata.values, df1['Litoral Leste'].values)

ax.legend(prop={'size':'11'})

plt.tight_layout()
```
